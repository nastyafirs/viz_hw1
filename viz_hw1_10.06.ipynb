```python
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import font_manager
from matplotlib.font_manager import FontProperties
```

```python
data = pd.read_csv('Pearson.tsv', sep='\t')
```

```python
print(data)
```

```python
x = data['Father']
y = data['Son']

plt.scatter(x,y)
```

```python
# font: https://www.fontsquirrel.com/fonts/FFF-Tusj

font = 'FFF_Tusj.TTF'
prop = font_manager.FontProperties(fname=font)

matplotlib.rcParams['figure.figsize'] = (13, 8)
grid = plt.GridSpec(15, 8, wspace=0.4, hspace=0.4)
general_ax = plt.subplot(grid[0:6, 0:4])

y_stats_ax1 = plt.subplot(grid[7:9, 0:4], xlabel= 'Seg. 1 on y axis')
x_stats_ax1 = plt.subplot(grid[0:6, 4:5], xlabel= 'Seg. 1 on x axis')
y_stats_ax2 = plt.subplot(grid[10:12, 0:4], xlabel= 'Seg. 2 on y axis')
x_stats_ax2 = plt.subplot(grid[0:6, 5:6], xlabel= 'Seg. 2 on x axis')
y_stats_ax3 = plt.subplot(grid[13:15, 0:4], xlabel= 'Seg. 3 on y axis')
x_stats_ax3 = plt.subplot(grid[0:6, 6:7], xlabel= 'Seg. 3 on x axis')

general_ax.scatter(x,y, color='green')

# Seg. 1 from 0 to 100
y_stats_ax1.hist(y[0:100], bins=50, color='purple')
y_stats_ax1.set_yticks([])
y_stats_ax1.invert_yaxis()
x_stats_ax1.hist(x[0:100], bins=50, orientation="horizontal", color='purple')
x_stats_ax1.invert_yaxis()
x_stats_ax1.set_yticks([])

# Seg. 2 from 200 to 450
y_stats_ax2.hist(y[200:450], bins=50, color='green')
y_stats_ax2.set_yticks([])
y_stats_ax2.invert_yaxis()
x_stats_ax2.hist(x[200:450], bins=50, orientation="horizontal", color='green')
x_stats_ax2.invert_yaxis()
x_stats_ax2.set_yticks([])

# Seg. 3 from 500 to 700
y_stats_ax3.hist(y[500:700], bins=50, color='purple')
y_stats_ax3.set_yticks([])
y_stats_ax3.invert_yaxis()
x_stats_ax3.hist(x[500:700], bins=50, orientation="horizontal", color='purple')
x_stats_ax3.invert_yaxis()
x_stats_ax3.set_yticks([])

figure = plt.gcf()
figure.suptitle("Pearson's father-son data analysis", fontproperties=prop, fontsize=30)
```
